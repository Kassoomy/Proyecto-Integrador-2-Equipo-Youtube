```
# Variable dependiente e independientes

X = df.drop(['marca_exito', 'Porcentaje de clics de las impresiones (%)'], axis= 1)
y = df.marca_exito
```

```
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss

def forward_feature_selection(df, target_column, num_features):
    # Lista para almacenar las características seleccionadas
    selected_features = []
    
    # Inicializar el modelo
    model = LogisticRegression()
    
    # Bucle para seleccionar características
    for _ in range(num_features):
        best_feature = None
        best_score = 0
        
        # Iterar sobre las características disponibles
        for feature in df.columns:
            # Verificar si la característica ya ha sido seleccionada
            if feature in selected_features:
                continue
            
            # Construir el conjunto de características actual
            feature_set = selected_features + [feature]
            
            
            # Entrenar y evaluar el modelo utilizando las características actuales
            model.fit(X, y)
            y_pred = model.predict(X)
            score = accuracy_score(y, y_pred) 
            ####ssr es más usado en modelos de regresión lineal, no en modelos de regresión logística. 
            #### Se debe a que la regresión logística es un modelo de clasificación y no de ajuste de valores continuos.
            #score = log_loss(y, y_pred) #log_loss: se encarga de calcular esta pérdida logarítmica.
            
            # Actualizar la mejor característica y su puntuación
            if score > best_score:
                best_feature = feature
                best_score = score
        
        # Agregar la mejor característica al conjunto seleccionado
        selected_features.append(best_feature)
    
    return selected_features

# Cargar el dataframe con 17 características (df) y especificar la columna objetivo (target_column)
# Asegúrate de que el dataframe esté correctamente preprocesado y listo para el análisis.

# Ejemplo de uso
num_features_to_select = 5  # Número de características a seleccionar
selected_features = forward_feature_selection(df, target_column, num_features_to_select)

print("Características seleccionadas:")
print(selected_features)
```

```
def scale(df):
    """Xi = Xi - mu(Xi) / σ(Xi)"""
    return (df-df.mean())/(df.std())
    
def forward_stepwise(X, y, results=[(0, [])]):
    """Perform forward stepwise variable selection as described in
    An Introductions to Statistical Learning (p.207)
    Returns pandas dataframe object  with results for each step"""
    # List predictors that havent's been used so far
    p_all    = list(X.columns)
    p_used   = results[-1][1]
    p_unused = [p for p in p_all if p not in p_used]
    
    # Job done, exit recursion
    if not p_unused:
        rss   = [r[0] for r in results]
        preds = [r[1] for r in results]
        return pd.DataFrame({'rss': rss, 'predictors': preds}).drop(0).reset_index()
    
    # Get rss score for each possible additional predictor
    r = []
    for p in p_unused:
        f     = 'marca_exito ~ {}'.format('+'.join([p]+p_used))
        # Fit model
        model = smf.ols(formula=f, data=pd.concat([X, y], axis=1)).fit()
        r     += [(model.ssr, [p]+p_used)]
    
    # Choose predictor which yields lowest rss
    min_rss     = sorted(r, key=lambda tup: tup[0])[0]   
    new_results = results + [min_rss]
    # Recursive call to self
    return forward_stepwise(X, y, new_results)
```

```
X = scale(X)

forward_stepwise_results = forward_stepwise(X, y)
display(HTML('<h4>Forward Stepwise Selection</h4>'))
display(forward_stepwise_results)
```

```
display(HTML('<h4>Comparison of methods for subset with 4 predictors</h4>'))
print('Forward Stepwise Selection  : {}'.format(sorted(forward_stepwise_results.loc[3]['predictors'])))
```

GPT: El método forward de selección de características se utiliza principalmente en modelos de aprendizaje supervisado, donde tienes una variable objetivo que deseas predecir. En este contexto, se evalúa el impacto de cada característica en la capacidad del modelo para realizar predicciones precisas.

El algoritmo K-means, por otro lado, es un algoritmo de aprendizaje no supervisado utilizado para agrupar datos en clusters. A diferencia de los modelos de aprendizaje supervisado, K-means no tiene una variable objetivo específica a predecir. En cambio, intenta encontrar agrupaciones naturales en los datos basándose únicamente en las características disponibles.

Dado que el método forward se basa en la evaluación del rendimiento del modelo utilizando la variable objetivo, no es aplicable directamente a modelos no supervisados como K-means.

```
def scale(df):
    """Xi = Xi - mu(Xi) / σ(Xi)"""
    return (df-df.mean())/(df.std())

def backward_stepwise(X, y, results=[]):
    """Returns pandas dataframe object  with results for each step"""
    # List predictors that havent's been used so far
    p_all    = list(X.columns)

    # Check if we're starting out here
    if not results:
        # Fit model with all features
        f     = 'marca_exito ~ {}'.format('+'.join(p_all))
        model = smf.ols(formula=f, data=pd.concat([X, y], axis=1)).fit()
        # Begin backward stepwise recursion
        return backward_stepwise(X, y, [(model.ssr, p_all)])
    else:
        p_used = results[-1][1]
    
    # Job done, exit recursion
    if len(p_used) == 1:
        rss   = [r[0] for r in results]
        preds = [r[1] for r in results]
        return pd.DataFrame({'rss': rss, 'predictors': preds})    
    
    # Get rss score for each possible removed predictor
    r = []
    for p in p_used:
        p_test = [i for i in p_used if i != p]
        f     = 'marca_exito ~ {}'.format('+'.join(p_test))
        # Fit model
        model = smf.ols(formula=f, data=pd.concat([X, y], axis=1)).fit()
        r     += [(model.ssr, p_test)]
    
    # Choose removal of predictor which yields lowest rss
    min_rss     = sorted(r, key=lambda tup: tup[0])[0]   
    new_results = results + [min_rss]
    return backward_stepwise(X, y, new_results)
```

```
X = scale(X)

backward_stepwise_results = backward_stepwise(X, y)
display(HTML('<h4>Backward Stepwise Selection</h4>'))
display(backward_stepwise_results)
```

```
display(HTML('<h4>Comparison of methods for subset with 4 predictors</h4>'))
print('Forward Stepwise Selection  : {}'.format(sorted(forward_stepwise_results.loc[3]['predictors'])))
print('Backward Stepwise Selection : {}'.format(sorted(backward_stepwise_results.loc[4]['predictors'])))
```
